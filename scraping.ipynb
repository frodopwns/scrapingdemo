```python
import requests
from time import sleep
from pprint import pprint
from urllib import unquote_plus

def get_next_target(page):
    start_link = page.find('<a href=')
    if start_link == -1: 
        return None, 0
    start_quote = page.find('"', start_link)
    end_quote = page.find('"', start_quote + 1)
    url = page[start_quote + 1:end_quote]
    return url, end_quote


def get_all_links(page):
    links = []
    while True:
        url, endpos = get_next_target(page)
        if url:
            links.append(url)
            page = page[endpos:]
        else:
            break
    return links


def get_section(text):
    #find 2nd table on page
    start = text.find("<table style=")
    #find its end
    stop = text.find("</table", start)

    return text[start: stop]
    

results = {}
base_url = "http://archive.freeteknomusic.org/"
r = requests.get(base_url)
    
if r.ok:
    data = get_all_links(get_section(r.text))
    #exclude links in first and last row -> data[3:-1]
    for u in data[3:6]:
        r = requests.get(base_url + u)
        links = get_all_links(get_section(r.text))
        u = u.lower()
        k = u[0]
        if k not in results:
            results[k] = {}
        results[k][u] = filter(lambda x: '.mp3' in x, links)
        sleep(1)

        
    for index in results:
        print index
        for band in results[index]:
            print "--", unquote_plus(band)
            for song in results[index][band]:
                print "  --", song
```

```python
from splinter import Browser
from BeautifulSoup import BeautifulSoup

def remove_pager(x):
    return x[0] != '' and x[0] != 'Page' and not x[0].startswith('Records')

def parse_source(html):
    soup = BeautifulSoup(html)
    table = soup.find("table")
    rows = table.findAll('tr')
    parsed = []
    for tr in rows[1:]:
        cols = tr.findAll('td')
        parsed.append([c.text for c in cols])
    parsed = filter(None, parsed)
    parsed = filter(remove_pager, parsed)
    return parsed

br = Browser()
br.visit("https://www.cusys.edu/budget/cusalaries/14.html")
print parse_source(br.html)
br.find_by_css("img[alt=\"Next\"]").click()
print parse_source(br.html)
#br.quit()
```
